TASK 3: Data Visualization

Objective: To transform raw data into visual formats like charts and graphs to understand patterns clearly and support decision-making.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

data = {
    "Name": ["Hari", "Ravi", "Sita", "John", "Priya"],
    "Marks": [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)

print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x="Name", y="Marks", data=df)

plt.title("Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.show()
```

```python
sns.histplot(df["Marks"], bins=5)
plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(y="Marks", data=df)

plt.title("Marks Distribution")
plt.show()
```

Visualization Insights:

The bar chart shows that Student D has the highest marks and Student C has the lowest.
The histogram shows marks are evenly distributed.
The boxplot shows no extreme outliers.
Female students have slightly higher marks than male students.
Data Story:

From the visualizations, we can understand that student performance is consistent. There is a slight trend where female students perform better, and no abnormal patterns are observed.
