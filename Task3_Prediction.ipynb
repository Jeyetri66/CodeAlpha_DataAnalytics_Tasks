TASK 4: Sentiment Analysis

Objective: To analyze text data from sources like product reviews and classify it as positive, negative, or neutral using NLP techniques, and understand public opinion and trends.

```python
from textblob import TextBlob
```

```python
reviews = [
    "This product is amazing and works perfectly",
    "Very bad experience, waste of money",
    "Average quality, not too good not too bad",
    "Excellent service and fast delivery",
    "Not satisfied with the product quality"
]

for review in reviews:
    analysis = TextBlob(review)

    if analysis.sentiment.polarity > 0:
        sentiment = "Positive"
    elif analysis.sentiment.polarity == 0:
        sentiment = "Neutral"
    else:
        sentiment = "Negative"

    print(review, "→", sentiment)
```

Sentiment Analysis Insights:

Positive reviews indicate customer satisfaction and good product performance.
Negative reviews highlight issues and dissatisfaction among customers.
Neutral reviews represent average or mixed opinions.
Public Opinion & Trends:

Most positive reviews suggest a good overall perception of the product.
Negative feedback helps identify areas for improvement.
Sentiment patterns help understand customer behavior and trends.
Applications:

Helps businesses in marketing strategies.
Supports product development and improvement.
Provides social insights from customer opinions.
